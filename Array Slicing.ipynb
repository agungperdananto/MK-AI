## Slicing Array

```python
x = np.arange(10,19)
x
```

```python
for i in range(9):
    print(x[i])
```

```python
for i in x:
    print(i)
```

```python
for i in range(9):
    if i>=4 and i<8 : 
        print(x[i])
```

```python
x = np.arange(15).reshape(3,5)
x  = x + 3.2
print(x)
```

```python
# get value from matrix x
for i in range (3):
    for j in range (5):
        print('data: ', x[i][j])
    
```

```python
for i in range (3):
    for j in range (5):
        if (i>=1) and ((j>=2) and(j<4)): 
            print('data: ', x[i][j])
```
